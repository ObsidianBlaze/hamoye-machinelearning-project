```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
# Getting the csv
car_csv_url = 'https://gist.githubusercontent.com/noamross/e5d3e859aa0c794be10b/raw/b999fb4425b54c63cab088c0ce2c0d6ce961a563/cars.csv' 
```

```
car_csv_url
```

```
carData = pd.read_csv(car_csv_url)
carData
```

```
# Plotting and giving the graph colors
carData.plot.scatter('hp','mpg', title = 'Plotting HP against MPG as a scatter plot', c='gear', colormap = 'viridis')
# Displaying the graph
plt.show()
```
