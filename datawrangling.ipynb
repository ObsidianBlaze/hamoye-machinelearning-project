**DATA WRANGLING**

```
import pandas as pd
```

Part One: Load Data


```
cellPhone_url = 'https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/cell_phones.csv'
population_url = 'https://raw.githubusercontent.com/zalihat/cell_phone_dataset/master/population1.csv'
geoCountry_url = 'https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv'
```

```
cellphoneData = pd.read_csv(cellPhone_url)
populationData = pd.read_csv(population_url)
geoData = pd.read_csv(geoCountry_url)
```

Part Two: Join Data


```
# Merging the cell phone and population on country and year
cellphoneMergePopulation = cellphoneData.merge(populationData, how='inner', on=['Country', 'year'])
cellphoneMergePopulation
```

```
# Renaming columns in the geo csv
geo_country = geoData.rename(columns={"country":"geo", "name": "Country"})
geo_country
```

```
# Merging the geo data with the merged cellphone country data.
country = geo_country[['Country','geo']]
country
```

```
# Merging geo population cell phonenumber and country
geo_population_cellphone_country_merge = country.merge(cellphoneMergePopulation, how='inner', on=['Country'])
```

Part Three: Make Features

```
geo_population_cellphone_country_merge
```

```
# Converting the population from int to float
print(geo_population_cellphone_country_merge['population'].dtype)

# Number of cell phones per person
# cellphone_sum = geo_population_cellphone_country_merge['cell_phones_total'].sum()
# population_sum = geo_population_cellphone_country_merge['population'].sum()

#dividing both sums.
adding = geo_population_cellphone_country_merge['cell_phones_total'] / geo_population_cellphone_country_merge['population']

# cellphone_per_person = cellphone_sum/population_sum

geo_population_cellphone_country_merge['total'] = adding

geo_population_cellphone_country_merge
```
